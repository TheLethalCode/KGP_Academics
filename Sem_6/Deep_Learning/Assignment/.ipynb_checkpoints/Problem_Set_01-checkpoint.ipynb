### Problem 1: Closed-Form Maximum Likelihood

Assume that we are given $n$ iid samples $(x_1, ..., x_n)$ from each $P(x \ | \ \theta)$ below. Compute the maximum likelihood estimates (MLEs) for the parameter $\theta$ of given distributions. 

**Q1.1**: $P(x \ | \ \theta) = \theta e^{-\theta x}$ for $x \geq 0$

**Q1.2**: $P(x \ | \ \theta) = \frac{1}{\theta}$ for $ 0 \leq x \leq \theta$

### Solution

Since $x_1, ..., x_n$ are iid samples,

&emsp; $\mathcal{L} (\theta ) = \mathcal{P} (x_1, ..., x_n ; \theta )$

&emsp;&emsp;&emsp;&emsp; $ = \displaystyle\prod_{i} \mathcal{P} (x_i ; \theta ) $

#### 1)

&emsp;&emsp; $\mathcal{L} (\theta ) = \displaystyle\prod_{i} \theta \cdot e^{-\theta x}$

&emsp;&emsp;&emsp;&emsp;&emsp;$ = \theta ^{n}\cdot e^{-\theta \cdot \displaystyle\sum_{i} x_i}$

&emsp;&nbsp; $ Taking \space log \space on \space both \space sides, $

&emsp;&emsp; $ \log{\mathcal{L} (\theta )} = n \cdot \log{\theta} - \theta \cdot \displaystyle\sum_{i} x_i$

&emsp;&nbsp; $ To \space maximize \space \mathcal{L}(\theta) , \space we \space can \space maximize \space \log{\mathcal{L} (\theta )} \space by \space equating \space its \space derivative \space w.r.t. \space \theta \space to \space 0. \space Differentiating \space w.r.t \space \theta, $

&emsp;&emsp; $ \frac{\displaystyle\partial \log{\mathcal{L} (\theta )} }{\displaystyle\partial \theta } = \frac{\displaystyle n}{\displaystyle \theta} - \displaystyle\sum_{i} x_i$

&emsp;&emsp;&emsp;&emsp; $ \frac{\displaystyle n}{\displaystyle\theta} - \displaystyle\sum_{i} x_i = 0$

&emsp;&emsp;&emsp;&emsp; $ \frac{\displaystyle n}{\displaystyle\theta} = \displaystyle\sum_{i} x_i $

### &emsp;&emsp; $ \boxed{\theta = \frac{n}{\displaystyle\sum_{i} x_i}}$

#### 2)

&emsp;&emsp; $\mathcal{L} (\theta ) = \displaystyle\prod_{i} \frac{1}{\theta}$

&emsp;&emsp;&emsp;&emsp;&emsp;$ = \theta^{-n}$

&emsp;&nbsp; $ We \space can \space see \space that \space to \space maximize \space \mathcal{L} (\theta ), \space \theta \space should \space be \space as \space low \space as \space possible. \space Since, \space 0 \leq x_i \leq \theta \space\space \forall x_i,$

### &ensp;&emsp; $ \boxed{\theta = \max_{i}{x_i}}$



### Problem 2: Gradient for Maximum Likelihood Estimation

Sometimes, we can't find MLEs in closed-form, in this case, we can use an iterative scheme to solve resulting optimization problem $\theta_{MLE} = \arg\max P(x \ | \ \theta)$. First-order iterative schemes (e.g. Gradient Descent, Stochastic Gradient Descent, Nesterov Accelerated Gradient Descent) require gradient information at each visited point. Compute gradients for log-likelihood of the following model:

- we have $X \in \mathbf R^{n \times d}$ - constant data matrix, $\mathbf x_i$ - vector corresponding to a single data point $(\mathbf x_i \in \mathbf R^{d})$
- $\theta$ is a $d$-dimensional (unknown) weight vector
- $\varepsilon \sim \text{Student}(v)$ is a $n$-dimensional (unknown) noise vector
- and we observe vector $\mathbf y = X\theta + \varepsilon$

Which means that
$$ P(y_i \ | \ \mathbf x_i, \theta, v) = \frac{1}{Z(v)} \Big(1 + \frac{(\theta^T \mathbf x_i - y_i)^2}{v}\Big)^{-\frac{v+1}{2}}$$

where $Z$ stands for a normalization constant that does not depend on $\theta$, and $v$ is a constant parameter.

Denote log-likelihood as $\mathcal L(\theta)$ and log-likelihood for a single sample $i$ as $\mathcal L_i(\theta) = \mathcal L(\theta \ | \ \mathbf x_i, y_i)$.

**Q2.1**: Find $\frac{\partial \mathcal L_i(\theta)}{\partial \theta}$.

#### Solution

$ We \space have, $

&emsp;&emsp;  $\mathcal L_i(\theta) = \mathcal L(\theta \ | \ \mathbf x_i, y_i)$

&emsp;&emsp;&emsp;&emsp;&emsp; $  = \log P(y_i \ | \ \mathbf x_i, \theta, v) $

&emsp;&emsp;&emsp;&emsp;&emsp; $  = \log{\Big(\frac{1}{Z(v)} \Big(1 + \frac{(\displaystyle\theta^T \mathbf x_i - y_i)^2}{\displaystyle v}\Big)^{-\frac{v+1}{2}}\Big )}$

&emsp;&emsp;&emsp;&emsp;&emsp; $  = \log{\frac{1}{Z(v)}} + \log{\Big( \Big( 1 + \frac{(\displaystyle \theta^T \mathbf x_i - y_i)^2}{\displaystyle v}\Big)^{-\frac{v+1}{2}}\Big )}$

&emsp;&emsp;&emsp;&emsp;&emsp; $  = -\log{Z(v)} - \frac{v+1}{2} \cdot \log{ \Big( 1 + \frac{(\displaystyle\theta^T \mathbf x_i - y_i)^2}{\displaystyle v}\Big)}$

$ Differentiating \space w.r.t \space \theta, $

&emsp;&emsp; $\frac{\displaystyle\partial \mathcal L_i(\theta)}{\displaystyle\partial \theta} = -\frac{\displaystyle\partial \log{Z(v)}}{\displaystyle\partial \theta} - \frac{\displaystyle v+1}{\displaystyle 2} \cdot \frac{\displaystyle\partial \log{ \Big( 1 + \frac{(\displaystyle\theta^T \mathbf x_i - y_i)^2}{\displaystyle v}\Big)}}{\displaystyle \partial \theta}$

&emsp;&emsp;&emsp;&emsp;&emsp;  $= -\frac{\displaystyle v+1}{\displaystyle 2} \cdot \frac{\displaystyle \partial \log{ \Big( 1 + \frac{(\displaystyle \theta^T \mathbf x_i - y_i)^2}{\displaystyle v}\Big)}}{\displaystyle \partial \theta}$    ${\ \ \ \ \ (v \ is \ constant \ w.r.t. \ to \ \theta)}$.

&emsp;&emsp;&emsp;&emsp;&emsp;  $= -\frac{\displaystyle v+1}{\displaystyle 2} \cdot \Big( 1 + \frac{(\displaystyle\theta^T \mathbf x_i - y_i)^2}{\displaystyle v}\Big)^{-1} \cdot \frac{\displaystyle\partial \Big( 1 + \frac{(\theta^T \mathbf x_i - y_i)^2}{v}\Big)}{\displaystyle\partial \theta}$

&emsp;&emsp;&emsp;&emsp;&emsp;  $= -\frac{\displaystyle v+1}{\displaystyle 2v} \cdot \Big( 1 + \frac{(\displaystyle\theta^T \mathbf x_i - y_i)^2}{v}\Big)^{-1} \cdot \frac{\displaystyle\partial \ (\theta^T \mathbf x_i - y_i)^2}{\displaystyle\partial \theta}$

&emsp;&emsp;&emsp;&emsp;&emsp;  $= -\frac{\displaystyle v+1}{\displaystyle v} \cdot \Big( 1 + \frac{(\displaystyle\theta^T \mathbf x_i - y_i)^2}{\displaystyle v}\Big)^{-1} \cdot (\displaystyle\theta^T \mathbf x_i - y_i) \cdot \frac{\displaystyle\partial \ (\theta^T \mathbf x_i - y_i)}{\displaystyle\partial \theta}$

&emsp;&emsp; $\boxed{\frac{\partial \mathcal L_i(\theta)}{\partial \theta}= -\frac{v+1}{v} \cdot \Big( 1 + \frac{(\theta^T \mathbf x_i - y_i)^2}{v}\Big)^{-1} \cdot (\theta^T \mathbf x_i - y_i) \cdot \mathbf x_i }$

**Q2.2**: Check the result obtained in the previous section by coding an SGD for $\mathcal L(\theta)$ and plotting how function value changes as a function of a number of passed gradient steps. Try multiple learning rates (1e-1, 1e-3, 1e-6), explain the result. Consider $Z(v)=1$ and $v=4$ for plotting.

We marked places where you are expected to add/change your own code with **`## -- ! code required`** comment.

```python
import numpy as np
import matplotlib.pyplot as plt


# a function to generate data
def generate_data(n=100, d=10, v=4):
    X = np.random.rand(n, d)*2-1
    theta = np.random.rand(d, 1)
    noise = np.random.standard_t(4, n).reshape((-1, 1))
    return X, X @ theta + noise


# the loss function which is the log-likelihood
def loss(X, y, theta, v=4):
    Z_v = 1
    loss = 0
    for i in range(X.shape[0]):
        e_i = (np.dot(X[i:i+1,:],theta) - y[i]).item()
        loss += - np.log(Z_v) - (v+1)/2 * np.log(1 + (e_i**2)/v)
    return loss


# the gradient of the negative of the loss function for a particular sample
def loss_grad_i(X, y, theta, i, v=4):
    
    e_i = (np.dot(X[i:i+1,:], theta) - y[i]).item()
    const = (v+1)/v * ((1 + (e_i**2)/v)**(-1)) * e_i
    return const * X[i:i+1,:].T


# function to plot the data
def sgd_plot(theta_init, X, y, n_steps=100):
    for learning_rate in [1e-1, 1e-3, 1e-6]:
        theta = theta_init.copy()
        losses = []
        for step_n in range(n_steps):
            for i in range(X.shape[0]):
                theta += -learning_rate * loss_grad_i(X, y, theta, i)
                loss_i = loss(X, y, theta)
                losses.append(loss_i)
        plt.plot(losses)
        plt.show()

        
def main(n=100, d=10, v=4):
    X, y = generate_data(n, d, v)
    theta_init = np.random.rand(d, 1)
    sgd_plot(theta_init, X, y)

    
main()
```

### Explanation

Here $ -\mathcal{L} (\theta) $ is being minimised with SGD, hence $ \mathcal{L} (\theta) $ is being maximized. From the maximum value of $ \mathcal{L} (\theta) $ we get $\theta _{MLE}$. Different learning rates approach the optimal value in different speeds. 

#### Learning rate 0.1:

We can see that the maximum value of $ \mathcal{L} (\theta) $ is reached around $100$ iterations, and after that the value is oscillating around the maximum. This happens because the step size is quite big, and hence the function crosses the maxima, after which it tries to come back to the maxima, when it again oversteps and crosses the maxima because of the high learning rate. This process repeats itself which can be seen as the oscillation of the value of $ \mathcal{L} (\theta) $

#### Learning rate 0.001:

In this case, $ \mathcal{L} (\theta) $ approaches its maximum value asymptotically, because of the optimal learning rate. As the function approaches its maximum value, its gradient decreases, and hence the change in $ \theta $ is also quite slow. By the end of $10000$ iterations, the function has almost reached its maximum value.

#### Learning rate $10^{-6}$:

As can be seen, in this case, $ \mathcal{L} (\theta) $ increases very slowly. Compared to the previous cases, it hasn't even reached $0.5 \% $ of the optimal value even after $10000$ iterations. This is because of the very low learning rate. 

### Problem 3: Logistic Regression

- and we observe $\mathbf y = \mathbf 1[X \theta + \varepsilon \geq 0]$ - boolean random vector

or, in other words: there exists some data $X$, we sample weights vector $\theta$ of the corresponding shape from multivariate normal distribution, multiply features from $X$ by weights from $\theta$, add random logistically distributed noise to resulting numbers, and observe whether resulting (noisy) random variable exceeds zero. So we are given the data matrix X and y, we want to estimate the weight vector $\theta$ as accurately as possible.

Do not get confused between standard [logistic function](https://en.wikipedia.org/wiki/Logistic_function#Mathematical_properties) and [Logistic](https://en.wikipedia.org/wiki/Logistic_distribution) distribution.

One way of estimating true weight vector $\theta$ is by finding an MLE estimate. To make use of prior information,  one might want to find [maximum a posteriori estimation](https://en.wikipedia.org/wiki/Maximum_a_posteriori_estimation) (MAP) that can be derived from likelihood $P(X|\theta)$ and prior $P(\theta)$ over weights as following (Bayes rule):

$$
\begin{align}
\theta_{MAP} &= \arg\max_\theta P(\theta | X) = \arg\max_\theta \frac{P(X|\theta)P(\theta)}{\int_{\Theta} P(X|\theta)P(\theta) d \theta} = \arg\max_\theta \big[ \log P(X|\theta) + \log P(\theta) - \log \int_{\Theta} P(X|\theta)P(\theta) d \theta \big] \\ &= \arg\max_\theta \big[ \log P(X|\theta) + \log P(\theta) \big] = \arg\max_\theta \mathcal L_{MAP}(\theta)
\end{align}
$$

In the model definition above, normal prior over weights is specified by line $\theta_j \sim \mathcal N(0, \sigma_\theta)$, which essentially means that "weight vector components should not be too far from zero".

**Q3.6**: $\frac{\partial \mathcal L_{MAP}(\theta)}{\partial \theta} = X^T(y - \text{logistic}(X\theta))) - 2 \lambda \theta $

#### Solution

&emsp;&emsp; $\frac{\displaystyle \partial \mathcal L_{MAP}(\theta)}{\displaystyle\partial \theta} = \frac{\displaystyle\displaystyle\partial \ y^T X \theta}{\displaystyle\partial \theta} - \frac{\displaystyle\partial \ \mathbf 1_{n \times 1}^T \cdot \log(\mathbf 1_{n \times 1} + \exp(X \theta))}{\displaystyle\displaystyle\partial \theta} - \frac{\displaystyle\displaystyle\displaystyle\partial \lambda \ ||\theta||^2_2}{\displaystyle\partial \theta} + \frac{\displaystyle\partial C}{\displaystyle\partial \theta}$


&emsp;&emsp; &emsp;&emsp; &emsp;&emsp; $ = (y^TX)^T - \Big(\mathbf 1_{n \times 1}^T \cdot \frac{\displaystyle\partial \log(\mathbf 1_{n \times 1} + \exp(X \theta))}{\displaystyle\displaystyle\partial \theta}\Big)^T - 2 \lambda \theta$

&emsp;&emsp; &emsp;&emsp; &emsp;&emsp; $ = (y^TX)^T - \Big(\mathbf 1_{n \times 1}^T \cdot \mathbf B\Big)^T - 2 \lambda \theta$ 

&emsp;&emsp; &emsp;&emsp; $ Here\ B_{ij} = \Big[\frac{\displaystyle\partial \log{(1+e^{\mathbf x_i\theta^T})}}{\displaystyle\partial \theta_j}\Big] $

&emsp;&emsp; &emsp;&emsp; &emsp; &emsp;&emsp;&emsp; &emsp;$ = \Big[\text{logistic}(\mathbf x_i\theta^T) \cdot x_{ij} \Big] $

&emsp;&emsp; &emsp;&emsp; $ (\mathbf 1_{n \times 1}^T \cdot \mathbf B)_j = \Big[\displaystyle\sum_i \text{logistic}(\mathbf x_i \theta^T) \cdot x_{ij}\Big]$

&emsp;&emsp; &emsp;&emsp; $ \mathbf 1_{n \times 1}^T \cdot \mathbf B = \text{logistic}(X\theta)^T \cdot X$

$ \Rightarrow \frac{\displaystyle \partial \mathcal L_{MAP}(\theta)}{\displaystyle\partial \theta} = (y^TX)^T - \Big(\text{logistic}(X\theta)^T \cdot X\Big)^T - 2 \lambda \theta$

$ \Rightarrow \boxed{\frac{\displaystyle \partial \mathcal L_{MAP}(\theta)}{\displaystyle\partial \theta} = X^T(y - \text{logistic}(X\theta)) - 2 \lambda \theta}$

Here's the code that generates data according to model specification (**Problem 3**) and plots for `k_features=2`.

```python
import numpy as np
import matplotlib.pyplot as plt


# A function to generate the data needed
def generate_logistic_regression_data(n_samples=200, k_featues=2, sigma_eps=0.1, 
                                      sigma_theta=1, feature_correlation_rate=0, train_test_ratio=4):
    
    assert 0 <= feature_correlation_rate <= 1
    
    # Create data points
    X_noncorr = np.random.rand(n_samples, k_featues)*2-1 # [0, 1] -> [-1, 1] center around zero
    mixture_matrix = np.eye(k_featues)*(1-feature_correlation_rate) + feature_correlation_rate/k_featues
    X = X_noncorr @ mixture_matrix
    
    # Generate theta according to normal distribution
    theta_real = np.random.normal(scale=sigma_theta, size=k_featues)[:, np.newaxis] # column vector
    
    # Generate noise according to logistic distribution
    noise = np.random.logistic(scale=sigma_eps, size=n_samples)[:, np.newaxis] # Column vector
    
    # Generate y from X, theta and noise
    y = ((X @ theta_real + noise) > 0).astype(np.int16)
    
    # Number of samples for training, splitting data into train and test
    split_idx = int(n_samples * train_test_ratio / (train_test_ratio + 1))
    
    # Return train X,Y, test X,Y, and theta
    return X[:split_idx, :], y[:split_idx], X[split_idx:, :], y[split_idx:], theta_real


# The function where the data is being plotted
def plot_2d_data(subfig, X, y, X_test, y_test, theta):
    
    assert len(X.shape) == 2
    
    # Plot the training data, with y = 0 points as red and y = 1 points as blue, marker = o
    subfig.scatter(X[:, 0], X[:, 1], color = [['r', 'b'][c] for c in np.ravel(y)], label="Training data")
    
    # Plot the test data, with marker changed to x
    point_color_arr = [['r', 'b'][c] for c in np.ravel(y_test)]
    subfig.scatter(X_test[:, 0], X_test[:, 1], color=point_color_arr, marker = 'x', label="Test data")
    
    # Plot the line represented by the theta vector
    subfig.plot([1, -1], [-theta[0]/theta[1], theta[0]/theta[1]])
    
    # Set the limit of the plots
    subfig.set_xlim(-1, 1)
    subfig.set_ylim(-1, 1)

    
# A generator function to generate the list of subplots along with the feature to be plotted
def subfig_iter(it, val_name=None, n_col=2, base_size=(8,8), title=None, fmt='%s'):
    
    # No of rows of plots. n_cols is the number of columns of the plot. In the output
    n_rows = len(it)//n_col + (len(it) % n_col != 0)
    
    # Size is the dimensions of the total plot. It increases with the number of rows and columns.
    # Maintains the size of a single subplot as constant
    size = (base_size[0]*n_col, base_size[1]*n_rows)
    
    # Create a grid of subplots. _axes corresponds to the particular subplots.
    fig, _axes = plt.subplots(n_rows, n_col, figsize=size)
    
    # Iterate through different features, assigning them to a particular subplot
    for subfig, val in zip(np.ravel(_axes), it):
        
        # Return the dict item and subfig as .__next__()
        # Wait for the data to be plotted
        yield (val, subfig)
        
        # Set the title
        if val_name is not None:
            if val_name == '':
                subtitle = fmt % (val)
            else:
                subtitle = ('%s = '+fmt) % (val_name, val)
            subfig.set_title(subtitle)
        
    # Place the legends for the whole figure in a single place
    # Get the handlers and labels
    handles, labels = np.ravel(_axes)[0].get_legend_handles_labels() 
    fig.legend(handles, labels, loc = 2) # Set the legend for the whole figure
    
    # Title for the total plots
    if title is not None:
        fig.suptitle(title) 


def test_draw():
    
    # Dictionary of various features
    data_args_dicts = [
        dict(), {'sigma_eps': 0.2}, {'sigma_eps': 0.5}, {'feature_correlation_rate': 0.3}, 
        {'feature_correlation_rate': 0.7}, {'feature_correlation_rate': 0.9}
    ]
    
    # Args for the generator function
    kwargs = {'val_name': 'property', 'title': 'PLOTS WITH DIFFERENT TYPE OF DATA'}
    
    # Iterate through the various features
    for arg_dict, subfig in subfig_iter(data_args_dicts, **kwargs):
        
        # Create necessary data
        data = generate_logistic_regression_data(**arg_dict)
        X_train, y_train, X_test, y_test, theta_real = data
        
        # Plot the data for that particular feature
        plot_2d_data(subfig, X_train, y_train, X_test, y_test, theta_real)
    
    # Display the plots
    plt.show()

    
test_draw()
```

**Q4.1.1**: When we use `feature_correlation_rate` ($\alpha$) the following happens: we construct a matrix $M$ that has values $1 - \frac{n-1}{n}\alpha$ on the diagonal and values $\frac{\alpha}{n}$ everywhere else. What happens to resulting matrix $XM$ if $\alpha$ is either 0 or 1? (If you do not know what feature correlation is here is a link that can help. [Feature Correlation](https://towardsdatascience.com/feature-selection-correlation-and-p-value-da8921bfb3cf))

#### Solution

Let $B = XM$

- $ \mathbf\alpha  = 0 $

When $\alpha = 0$, the diagonal elements are $1$, and $0$ elsewhere. This means, there is no correlation between the features. Each element $B_{ij}$ is a randomly generated number, as $B_{ij} = 1 \cdot X_{i1} + 0 \cdot X_{i2}$ or $B_{ij} = 0 \cdot X_{i1} + 1 \cdot X_{i2}$, and $X_{ij}$ are randomly generated.

- $ \mathbf\alpha  = 1 $

When $\alpha = 1$, all elements are $\frac{1}{2}$ everywhere. This means, two features of the generated data are positively linearly dependent on each other. Here, $B_{ij} = 1 \cdot X_{i1} + 1 \cdot X_{i2}$, and $X_{ij}$ are randomly generated. This means that $B_{i1} = B_{i2}$. So, if the correlation feature is 1, the two features of the generated data end up being the same.

**Q4.2**: Why did we implement $\text{softplus}(x) = \log(1+\exp(x))$ as shown below? What is numerical under- and overflow and how people deal with it?

#### Solution

##### Numeric Overflow

When a number is larger than what can be represented by the number of bits alloted by the program, it will result in something called Numeric Overflow. In Python, in case of Floating point values, when they exceed a particular value, they result in Overflow.

##### Numeric Underflow

During a computation, when a number is so close to zero that precision was lost, it will result in something called Numeric Underflow. In Python, in case of Floating point values, when they go below a particular value, they result in Underflow.

##### Handling Overflow and Underflow

In case numeric overflows and underflows are possible during a particular computation, it is better to approximate the result for which these errors might occur. These are quite common when exponentiation is carried out. For example, take $ \text{softplus}(x) = \log(1 + \exp(x))$ implemented in the below code:-

The function is implemented in such a way, so that it isn't affected by numerical overflow and underflow. Let's look at the behavior of the $\text{softplus}(x)$.

- For $x \geq \text{limit}, \log(1+\text{exp}(x)) \approx \log(\text{exp}(x)) = x$
- For $x \leq \text{-limit}, \log(1+\text{exp}(x)) \approx \text{exp}(x) \approx 0$
- For $\text{-limit} < x < \text{limit}, \log(1+\text{exp}(x))$

Here the softplus function is approximated outside a certain range because, for big $x$, computation of $\text{exp}(x)$ may result in $\text{Overflow}$, and for big $-x$, it may result in $\text{Underflow}$. The above approximation ensures that these do not occur during the calculation of the $\text{softplus}(x).$  

```python
import numpy as np


def softplus(x_, limit=5):

    x = np.array(x_) # Convert to numpy array
    
    # Set value True for those elements that satisfy -limit < x[i] < limit
    compute_real_mask = np.logical_and(-limit < x, x < limit)
    
    # Set value True for those elements that satisfy x[i] >= limit
    return_same_mask = x >= limit
    
    # Elements where |x[i]| >= limit will be 0 
    computed_real_part = np.log(1 + np.exp(x*compute_real_mask)*compute_real_mask)
    
    # Have elements where x[i] >= limit will be retained as it is
    returned_same_part = x*return_same_mask
    
    # Combine both. This means all elements with x[i] <= -5 will become zero
    return computed_real_part + returned_same_part


def test_softplus():
    
    x_arr = np.linspace(-200, 200) # A sample of 50 evenly spaced data
    softplus_true = np.log(1 + np.exp(x_arr)) # Calculate actual softplus of data
    
    # S[i] = 0, where x[i] <= -limit.
    # S[i] = actual_softplus, where -limit < x[i] < limit.
    # S[i] = x[i], where x[i] >= limit.
    softplus_stable = softplus(x_arr)
    
    # Makes sure the error of softplus stable is within acceptable range
    assert np.max(abs(softplus_true-softplus_stable)) < 1e-3

    
test_softplus()

def logistic(x):
    return 1/(1+np.exp(-x))
```

**Q4.3**: Write (by filling in the missing code below) Gradient Descent and Stochastic Gradient Descent algorithm for the MAP loss and plot 

- MAP loss
- L2 distance between current theta and true one
- cosine distance between current theta and true one

use results from **Q3.6**. Note that our loss is $loss(\theta) = -\mathcal L(\theta)$ and it is usually referred as "the negative log-likelihood".

```python
from numpy.linalg import norm
from scipy.spatial.distance import cosine
import matplotlib.cm as cm


# loss(theta) = -L(theta)
def logistic_MAP_loss(X, y, theta_current, lambd):
    
    # Calculate the different components of the loss
    loss1 = (y.T @ X @ theta_current).item() 
    loss2 = (np.ones((1, X.shape[0])) @ np.log(1 + np.exp(X @ theta_current))).item()
    loss3 = lambd * (norm(theta_current[:,0])**2)

    # Merge the losses
    return -(loss1-loss2-loss3)


# gradient of loss(theta)
def logistic_MAP_loss_gradient(X, y, theta_current, lambd):
    
    # Different components of loss
    loss1 = X.T @ (y - logistic(X @ theta_current))
    loss2 = 2 * lambd * theta_current
    
    # Merge them
    return -(loss1 - loss2)


# Gradient considering only a single sample
def logistic_MAP_loss_single_gradient(X, y, smaple_id, theta_current, lambd):
    
    x_i = X[smaple_id].reshape((1, -1)) # Transforms into row vector
    y_i = y[smaple_id].reshape((1, 1)) # Transforms into column vector
    return logistic_MAP_loss_gradient(x_i, y_i, theta_current, lambd)


# The accuracy of our classification for the test data
def logistic_classification_accuracy(X, y, theta_current):
    
    y_calc = ((X @ theta_current) > 0).astype(np.int16)
    return np.mean(y_calc == y)
    

# A Gradient Descent approach
def run_MAP_gd(X, y, lambd, learning_rate = 0.01, max_epoch_n=200, init_theta=None):
    
    # Initialise the weight vector
    theta = init_theta if init_theta is not None else np.ones((X.shape[1], 1))*5
    steps = [theta.copy()]
    
    # Run the number of iterations needed
    for epoch_n in range(max_epoch_n):
        theta += -learning_rate * logistic_MAP_loss_gradient(X, y, theta, lambd) # Update theta
        steps.append(theta.copy())  # Append theta to the list
        
    return steps


# A Stochastic Gradient Descent approach
def run_MAP_sgd(X, y, lambd, learning_rate = 0.01, max_epoch_n=200, init_theta=None):
    
    # Set initial value of theta as a column vector of same values
    theta = init_theta if init_theta is not None else np.ones((X.shape[1], 1))*5
    steps = [theta.copy()]
    
    # Run the number of iterations
    for epoch_n in range(max_epoch_n):
        for sample_id in range(X.shape[0]):
            # Update theta
            theta += -learning_rate * logistic_MAP_loss_single_gradient(X, y, sample_id, theta, lambd)
            
        steps.append(theta.copy()) # Append theta to the list
        
    return steps
```

```python
import time


def draw_gd_plots(axes, theta_steps, X, y, lambd, theta_real, label_plot):
    
    collected_val_dicts = []
    
    # Go through all the in between values of theta
    for theta in theta_steps:
        collected_val_dicts.append({
                'distance_to_true': np.linalg.norm(theta-theta_real),  # Actual distance 
                'cosine_dist_to_true': cosine(theta, theta_real),      # Cosine distance
                'map_loss': float(logistic_MAP_loss(X, y, theta, lambd)), # The Loss Function (negative log likelihood)
            })
    
    # Plot the values for each of the subplots
    for plot_id, key in enumerate(collected_val_dicts[0].keys()):
        subfig = axes[plot_id]
        subfig.set_title(key)
        subfig.plot([d[key] for d in collected_val_dicts], label=label_plot)
        subfig.legend(loc=0)

        
def main():
    
    # No. of features = 2 and 100
    # Learning rate = 1e-3 and 5*1e-1
    for k_featues in [2, 100]:
        for learning_rate in [1e-3, 0.9]:
            
            # Generate data with the given args, with features having no correlation
            data_args = {'n_samples': 200, 'k_featues': k_featues, 'sigma_eps': 0.3, 
                         'sigma_theta': 5, 'feature_correlation_rate': 0}
            data = generate_logistic_regression_data(**data_args)
            X_train, y_train, X_test, y_test, theta_real = data
            lamdb = 0

            # Three different subplots
            fig, axes = plt.subplots(1, 3, figsize=(18,6)) 

            ## Gradient Descent approach (Blue Plot)
            now = time.process_time() # Start the timer
            steps = run_MAP_gd(X_train, y_train, lamdb, learning_rate=learning_rate, max_epoch_n=1000) # Run a GD
            draw_gd_plots(axes, steps, X_train, y_train, lamdb, theta_real, 'GD') # Draw the plots
            end1 = time.process_time() # Running time calculations
            accuracy = logistic_classification_accuracy(X_test, y_test, steps[-1]) # Calculate the accuracy on the test set
            print('GD runtime:', end1 - now)
            print('GD accuracy: %.3f' % (accuracy * 100))


            ## Stochastic Gradient Descent approach (Orange Plot)
            now = time.process_time() # Start the timer
            steps = run_MAP_sgd(X_train, y_train, lamdb, learning_rate=learning_rate, max_epoch_n=1000) # Run a SGD
            draw_gd_plots(axes, steps, X_train, y_train, lamdb, theta_real, 'SGD') # Draw the plots
            end1 = time.process_time() # Running time calculations
            accuracy = logistic_classification_accuracy(X_test, y_test, steps[-1]) # Calculate the accuracy on the test set
            print('SGD runtime:', end1 - now)
            print('SGD accuracy: %.3f' % (accuracy * 100))


            fig.suptitle('GD vs SGD, k_featues = %d, learning_rate = %.3f' % (k_featues, learning_rate), verticalalignment='bottom') # Set the title of the plots
            plt.show() # Display the plots

        
main()
```

**Q4.3.1**: We expect loss to gradually decrease. Re-run code above several times (sgd might work differently). Explain plots above.

#### Solution

##### The 3 Plots

- The first plot compares "$\textbf{how far}$" is calculated $\theta_i$ compared to the actual $\theta$. The value of the plot decreases as $\theta_i$ approaches the actual $\theta$. Note that, this graph need not always be decreasing. This is because what we are trying to minimize is $-\mathcal{L}(\theta)$, and while the actual $\theta$  evaluates to a high $\mathcal{L}(\theta)$, owing to the probabilistic nature of the likelihood function and $\theta$ being a random variable, there might be a value of $\theta_i$ for which $\mathcal{L}(\theta_i)$ evaluates higher. Hence, this graph, in general, has a local minima.

- In the second plot, we try to observe the "$\textbf{orientation}$" of the calculated $\theta_i$ with respect to the actual $\theta$. Here, a lesser value implies the angle between $\theta_i$ and $\theta$ is small. Only when the directions of the two vectors are similar, the loss function is minimized. $$ \text{cosine}(\theta, \theta_i) = 1 - \frac{\theta \cdot \theta_i}{||\theta||_2\cdot||\theta_i||_2} $$

- The final graph depicts the change in the "$\textbf{loss function}(-\mathcal{L}(\theta))$ " as we try to minimize it. Since both GD and SGD is all about minimizing the loss function, we can see a steady decrease in the loss function as $\theta_i$ approaches the optimal value. As the function approaches its final value, we can see there is a decrease in the rate of its decrease. Hence, as the number of epochs increase, the slope of the curve decreases.


##### SGD vs GD

In the graphs above, both SGD and GD are plotted, but since $\theta_i$ in both cases approach the optimal value quite similarly, the difference between the graphs is not obvious and they overlap. This mainly occurs because, $\theta_i$ is compared every epoch. If every update of $\theta_i$ in SGD is plotted, we can see its jittery approach towards the optimum. Another factor is the low learning rate. The low learning rate suppresses any erratic updates that might occur in the SGD making it so that it behaves like GD. When the learning rate is increased, we can see the graphs diverge.

##### Runtime

As we can see, the runtimes for both SGD and GD do not change drastically with the number of features, owing to the vectorized calculations. Even for a $5000 \% $ increase in the feature dimensions, the increase in runtime is just $134.5\%$ and $110.3 \%$ for GD and SGD, respectively.

The runtimes for SGD and GD also differ exactly because of this reason. In each epoch in GD, the gradient is found as a function of the matrix of the entire data points, which optimizes it. But in SGD, in each epoch, every point is individually considered, and gradient is found. This slows down the calculations, which can be seen in the increased runtimes in both cases for SGD.
